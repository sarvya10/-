```python
cust_acqu = pd.read_csv(r"/content/Cust_acqu.csv")
cust_acqu.pop("Sl No:")
cust_acqu
```

```python
spend = spend.rename(columns=({"Costomer":"Customer"}))
spend.pop("Sl No:")
spend["Month"]=spend["Month"].apply(lambda x:pd.to_datetime(x))
spend
```

```python
repay = pd.read_csv(r"/content/Repay.csv")
repay
```

```python
repay = repay.rename(columns=({"Costomer":"Customer"}))
repay.pop("SL No:")
repay["Month"]=repay["Month"].apply(lambda x:pd.to_datetime(x))
repay
```

# Sanity Checks – Data Cleaning

### Prob. 1 : Provide a meaningful treatment to all values where age is less than 18.

```python
cust_acqu.loc[(cust_acqu.Age<18),"Age"]=cust_acqu.Age.mean()
cust_acqu
```

### Prob. 2 : Identity where the repayment is more than the spend then give them a credit of 2% of their credit limit in the next month billing.

```python
cust_repay = pd.merge(left=spend,right=repay,on=["Customer","Month"])
cust_repay
```

```python
cust_repay["Identify_repayment"]=cust_repay.Amount_y>cust_repay.Amount_x
cust_repay
```

```python
cust_repay["surplus"]=cust_repay.Amount_y - cust_repay.Amount_x
cust_repay["Bonus"]=np.where(cust_repay["surplus"]>0,cust_repay["surplus"]*2/100,np.NaN)
cust_repay
```

# Tasks

### Prob. 1 : Monthly spend of each customer.

```python
cust_spend = pd.merge(left=cust_acqu,right=spend,on="Customer")
cust_spend
```

```python
cust_spend["Monthly_Spend"]=cust_spend["Month"].apply(lambda x:x.month)
monthly_spend=cust_spend.groupby(["Customer","Monthly_Spend"])["Amount"].sum().reset_index()
monthly_spend
```

### Prob. 2 : Monthly repayment of each customer.

```python
cust_repay =pd.merge(left=cust_acqu,right=repay,on="Customer")
cust_repay
```

```python
cust_repay["Monthly_Spend"]=cust_repay["Month"].apply(lambda x:x.month)
monthly_repay=cust_repay.groupby(["Customer","Monthly_Spend"])["Amount"].sum().reset_index()
monthly_repay
```

### Prob. 3 : Highest paying 10 customers.

```python
spend_bonus=pd.merge(left=cust_acqu,right=spend,on="Customer")
spend_bonus
```

```python
spend_bonus.groupby("Customer").Amount.sum().reset_index().sort_values("Amount",ascending=False)
```

### Prob. 4 :  People in which segment are spending more money.

```python
spend_bonus.groupby("Segment").Amount.sum().reset_index().sort_values("Amount",ascending=False)
```

### Prob. 5 :  Which age group is spending more money?

```python
spend_bonus["Age_Group"]=np.where(((spend_bonus.Age.astype(int)>=18) & (spend_bonus.Age.astype(int)<25)), "Youth",np.where(((spend_bonus.Age.astype(int)>=25) & (spend_bonus.Age.astype(int)<65)), "Adult","Senior"))
spend_bonus.groupby("Age_Group").Amount.sum()
```

### Prob. 6 :  Which is the most profitable segment?

```python
cust_trans=pd.merge(left=cust_repay,right=spend,on="Customer")
cust_trans
```

### Prob. 7 : In which category the customers are spending more money?

```python
spend_bonus.groupby("Type").Amount.sum().reset_index().sort_values("Amount",ascending=False)
```

### Prob. 8 : Monthly profit for the bank.

```python
monthly_profit = pd.merge(left=spend,right=repay,on=["Customer","Month"])
monthly_profit["Monthly_Profit"]=monthly_profit.Amount_y-monthly_profit.Amount_x
monthly_profit
```

### Prob. 9 :  Impose an interest rate of 2.9% for each customer for any due amount.

```python
due = pd.merge(left=cust_acqu,right=spend,on=["Customer"])
cust_due=pd.merge(left=due,right=repay,on=["Customer","Month"])
cust_due
```

```python
cust_due["Remaing_payment"]=cust_due.Amount_x-cust_due.Amount_y
cust_due["Due"]=np.where(cust_due["Remaing_payment"]>0,cust_due["Remaing_payment"]*2.9/100,np.NaN)
cust_due
```
